```python
# file_name = "homography0.csv"
file_name = "friday0_frm.csv"
data = np.loadtxt(file_name, skiprows=1, delimiter=',')
data.shape
homo_x = data[:,2]
homo_y = data[:,3]
cam_x = data[:,4]
cam_y = data[:,5]-1
L2 = data[:,7]
ids = data[:,8]

blue, orange = '#1f77b4', '#ff7f0e'

max_id = np.max(ids)
plt.figure(figsize=(10,6))

plt.title("XY plot")
for i in range(int(max_id+1)):
    idxs = np.where(ids ==i)
    if i == 0:
        plt.plot(cam_x[idxs],cam_y[idxs],color ="C1", label="Ground truth")
        plt.plot(homo_x[idxs],homo_y[idxs],"--",color="C0", label="Homography")
        
    else:
        plt.plot(cam_x[idxs],cam_y[idxs],color ="C1", label="")
        plt.plot(homo_x[idxs],homo_y[idxs],"--",color="C0", label="")
        

# ax = plt.gca()
# ax.invert_xaxis()
# ax.invert_yaxis()

plt.xlabel("x[cm]")
plt.ylabel("y[cm]")
plt.legend(loc='upper right')
plt.show()


plt.figure(figsize=(10,6))
plt.title("Euclidean distance")
plt.plot(L2)
plt.xlabel("Iter[-]")
plt.ylabel("L2 Norm[cm]") # Euclid dist, ...
plt.show()

plt.title("Distances in axis ")
plt.plot(L2_x,label="Distance in x")
plt.plot(L2_y,"--",label="Distance in y")
plt.xlabel("Iter[-]")
plt.ylabel("L2 Norm[cm]") # Euclid dist, ...
plt.legend()
plt.show()

# plt.title("Distances in axis")
# plt.plot(np.abs(data[:,2]-data[:,4]),label="Distance in x")
# plt.plot(np.abs(data[:,3]-data[:,5]),"--",label="Distance in y")
# plt.xlabel("Iter[-]")
# plt.ylabel("L2 Norm[mm]") # Euclid dist, ...
# plt.legend()
# plt.show()

# plt.title("ENCODER CORPARISON")
# plt.plot(np.abs(data[:,1] - data[0,1]), label="Encoder")
# plt.plot(np.abs(data[:,2]-data[0,2]),label="Homography")
# plt.plot(np.abs(data[:,4]-data[0,4]),label="Camera")
# plt.xlabel("Iter[-]")
# plt.ylabel("Diff to start[mm]")
# plt.legend()
# plt.show()

# plt.title("DIFF2ENCODER")
# # plt.plot(np.abs(data[:,1] - data[0,1]), label="Encoder")
# plt.plot(np.abs(np.abs(data[:,2]-data[0,2])-np.abs(data[:,1] - data[0,1])),label="Homography")
# plt.plot(np.abs(np.abs(data[:,4]-data[0,4])-np.abs(data[:,1] - data[0,1])),label="Camera")
# plt.xlabel("Iter[-]")
# plt.ylabel("Diff to start[mm]")
# plt.legend()
# plt.show()
```
